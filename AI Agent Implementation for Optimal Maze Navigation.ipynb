* **Deterministic** means that the **transition function contains no randomness.** An action in a state will always produce the same result. Going south from the start state always will lead to the same square.
* **Know** means that the agent **knows the complete transition function.** The
agent has the map and therefore knows how its position changes when it walks in a direction.

## Tree structure

Here is an implementation of the basic node structure for the search algorithms (see Fig 3.7 on page 73). I have added a method that extracts the path from the root node to the current node. It can be used to get the path when the search is completed.

```python
class Node:
    def __init__(self, pos, parent, action, cost):
        self.pos = tuple(pos)    # the state; positions are (row,col)
        self.parent = parent     # reference to parent node. None means root node.
        self.action = action     # action used in the transition function (root node has None)
        self.cost = cost         # for uniform cost this is the depth. It is also g(n) for A* search

    def __str__(self):
        return f"Node - pos = {self.pos}; action = {self.action}; cost = {self.cost}"

    def get_path_from_root(self):
        """returns nodes on the path from the root to the current node."""
        node = self
        path = [node]

        while not node.parent is None:
            node = node.parent
            path.append(node)

        path.reverse()

        return(path)
```

### How does BFS and DFS (without a reached data structure) deal with loops (cycles)?

Managing cycles or loops is essential to maintaining efficiency and avoiding infinite loops in maze-solving algorithms such as Depth-First Search (DFS) and Breadth-First Search (BFS). BFS explores nodes level by level by using a queue. It tracks positions that have previously been investigated by keeping a visited set. In order to avoid needless re-exploration and infinite loops, BFS bypasses nodes that it detects as having been visited. By using this method, BFS can make sure that each node is only processed once as it proceeds through the maze methodically. On the other hand, DFS uses a stack to traverse, going as far down each branch as it can before going back. DFS keeps track of a visited set similarly to BFS.

Upon returning to a node that has previously been designated as visited, DFS goes back to the initial node in order to prevent loops. Nevertheless, if DFS is not handled properly, it may become stuck since it descends further into routes. DFS might perpetually loop between two or more connected nodes if it lacked an appropriate way to track visited nodes. In conclusion, a visited set—which forbids revisiting nodes—is used by both BFS and DFS to efficiently handle loops. Whereas DFS needs to go backward in order to prevent cycles, BFS's level-wise search naturally reduces that danger. Although the traversal strategies of the two techniques differ, they both stress the significance of monitoring visited nodes to guarantee accuracy and efficiency in maze navigation.

### Are your implementations complete and optimal? Explain why. What is the time and space complexity of each of **your** implementations? Especially discuss the difference in space complexity between BFS and DFS.

The given code includes implementations of Depth-First Search (DFS) and Breadth-First Search (BFS) that are capable of handling a variety of circumstances and efficiently navigate mazes. To make sure it discovers the shortest path in unweighted graphs, BFS methodically investigates every nearby node at the current depth before going deeper. Its space complexity is also O(b^d) since it keeps all nodes at the current level in a queue, which makes it less memory-efficient, especially in big or complex mazes. Its time complexity is 𝑂(𝑏^𝑑), where b is the branching factor and d is the depth of the shallowest solution. DFS, on the other hand, takes a different tack and explores as far down a branch as feasible before turning around.

Since it only keeps the current path and its immediate siblings, its space complexity is lowered to O(b⋅d), making it more memory-efficient in deep, narrow trees, even if it still has the same time complexity of O(b^d). Both algorithms are capable of finding solutions and are complete; but, depending on the particular problem setting, their efficacy differs. In scenarios where memory limitations are crucial, DFS might be more appropriate, but BFS is superior when the objective is to discover the shortest path. Choosing the right algorithm for the maze or search problem at hand depends on your understanding of these trade-offs.

```python
# Sample maze utility functions
def parse_maze(maze_content):
    return [list(line.replace(" ", "")) for line in maze_content.strip().split('\n')]

def show_maze(maze):
    for row in maze:
        print(' '.join(row))
    print()

def find_pos(maze, what="S"):
    for r, row in enumerate(maze):
        for c, cell in enumerate(row):
            if cell == what:
                return (r, c)
    return None

class PathNode:
    def __init__(self, position, predecessor, action_taken, total_cost):
        self.position = tuple(position)
        self.predecessor = predecessor
        self.action_taken = action_taken
        self.total_cost = total_cost

    def retrieve_path_from_root(self):
        current_node = self
        path_sequence = [current_node]

        while current_node.predecessor is not None:
            current_node = current_node.predecessor
            path_sequence.append(current_node)

        path_sequence.reverse()
        return path_sequence

def is_valid_move(maze, position):
    rows, cols = len(maze), len(maze[0])
    r, c = position
    return 0 <= r < rows and 0 <= c < cols and maze[r][c] != "X"

def breadth_first_search(maze, start_pos, goal_pos, movements, initial_node):
    explored = set()
    search_queue = [initial_node]

    while search_queue:
        current_node = search_queue.pop(0)
        current_location = current_node.position

        if current_location in goal_pos:
            return current_node

        explored.add(current_location)

        for move in movements.values():
            new_location = (current_location[0] + move[0], current_location[1] + move[1])
            if is_valid_move(maze, new_location) and new_location not in explored:
                child_node = PathNode(new_location, current_node, move, current_node.total_cost + 1)
                search_queue.append(child_node)

    return None

def depth_first_search(maze, start_pos, goal_pos, movements, initial_node, depth_limit):
    explored = set()
    search_stack = [(initial_node, 0)]

    while search_stack:
        current_node, depth = search_stack.pop()
        current_location = current_node.position

        if current_location in goal_pos:
            return current_node

        if depth < depth_limit:
            explored.add(current_location)

            for move in movements.values():
                new_location = (current_location[0] + move[0], current_location[1] + move[1])
                if is_valid_move(maze, new_location) and new_location not in explored:
                    child_node = PathNode(new_location, current_node, move, current_node.total_cost + 1)
                    search_stack.append((child_node, depth + 1))

    return None

def iterative_deepening_search(maze, start_pos, goal_pos, movements, initial_node):
    depth_limit = 0
    while True:
        print(f"Searching with depth limit: {depth_limit}")
        result = depth_first_search(maze, start_pos, goal_pos, movements, initial_node, depth_limit)
        if result:
            return result
        depth_limit += 1

def performSearch(maze, searchMethod="BFS"):
    startPos = find_pos(maze, "S")
    goalPos = [find_pos(maze, "G1"), find_pos(maze, "G2")]

    directions = {
        "north": (-1, 0),
        "south": (1, 0),
        "east": (0, 1),
        "west": (0, -1)
    }

    initialNode = PathNode(startPos, None, None, 0)

    if searchMethod.upper() == "BFS":
        endNode = breadth_first_search(maze, startPos, goalPos, directions, initialNode)
    elif searchMethod.upper() == "DFS":
        endNode = depth_first_search(maze, startPos, goalPos, directions, initialNode, depth_limit=10)
    elif searchMethod.upper() == "IDS":
        endNode = iterative_deepening_search(maze, startPos, goalPos, directions, initialNode)
    else:
        print("Search method not implemented.")

    if endNode:
        path = endNode.retrieve_path_from_root()
        print("Path from start to goal:")
        for node in path:
            print(node.position)
            if node.position in goalPos:
                maze[node.position[0]][node.position[1]] = "P"
        print("Cost:", endNode.total_cost)
    else:
        print("Goal not found")
    show_maze(maze)

def runMazes():
    # Medium maze with multiple goals
    maze_content = """
S . . X . G1
. X . X . .
. . . . X G2
X X . . . .
. . X X X .
    """
    maze = parse_maze(maze_content)
    show_maze(maze)

    print("Running BFS:")
    performSearch(maze, "BFS")
    
    maze = parse_maze(maze_content)  # Reset maze
    print("Running DFS:")
    performSearch(maze, "DFS")
    
    maze = parse_maze(maze_content)  # Reset maze
    print("Running IDS:")
    performSearch(maze, "IDS")

runMazes()
```
